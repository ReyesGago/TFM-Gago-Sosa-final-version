# Count Terms

```python
import csv
import re

# Function to count exact occurrences of a term in a text file
def count_term_occurrences(term, text_file):
    """
    Counts the exact occurrences of a term in a text file.

    Args:
        term (str): The term to search for.
        text_file (str): Path to the text file.

    Returns:
        int: The number of occurrences of the term in the text file.
    """
    pattern = rf"\b{re.escape(term)}\b"  # \b ensures it matches the exact word
    count = 0

    with open(text_file, 'r', encoding='utf-8') as file:
        for line in file:  # Process line by line
            count += len(re.findall(pattern, line, re.IGNORECASE))  # Count occurrences in the line

    return count

# Function to extract unique terms from the first column of a CSV file
def extract_unique_terms(csv_file_path):
    """
    Extracts unique terms from the first column of a CSV file.

    Args:
        csv_file_path (str): Path to the CSV file.

    Returns:
        set: A set of unique terms.
    """
    unique_terms = set()
    with open(csv_file_path, mode='r', encoding='utf-8') as file:
        reader = csv.reader(file)
        next(reader)  # Skip the header row
        for row in reader:
            if row:
                unique_terms.add(row[0].strip())
    return unique_terms

# Main function to count term occurrences and save results to an output file
def process_terms_and_count(csv_file_path, text_file_path, output_file_path):
    """
    Reads terms from a CSV file, counts their occurrences in a text file, 
    and writes the results to an output file.

    Args:
        csv_file_path (str): Path to the CSV file containing terms.
        text_file_path (str): Path to the text file to search for terms.
        output_file_path (str): Path to the output file to save results.
    """
    unique_terms = extract_unique_terms(csv_file_path)

    with open(output_file_path, 'w', encoding='utf-8') as output_file:
        for term in unique_terms:
            occurrences = count_term_occurrences(term, text_file_path)
            output_file.write(f"{term}: appears {occurrences} time(s).\n")
            print(f"✅ {term}: {occurrences} time(s) in the corpus" if occurrences > 0 else f"❌ {term}: Does not appear in the corpus")

# File paths
csv_file_path = "basque/BA-EN-[Medical Sciences ~ Pharmaceutical Technology]_and_[Technological Sciences ~ Environmental technology and engineering].csv"
text_file_path = "basque/depth_4/specific_wikipedia_files/output_ba_environmental.txt"
output_file_path = "basque/depth_4/terms_counter/terms_number_environmental_ba_en_v2.txt"

# Execute the process
process_terms_and_count(csv_file_path, text_file_path, output_file_path)
```

### Make a comparison with graphs


```python
import re
import matplotlib.pyplot as plt
import pandas as pd

# Function to read a text file and extract terms and their frequencies into a dictionary
def read_terms_from_file(file_path):
    """
    Reads a text file and extracts terms and their frequencies.

    Args:
        file_path (str): Path to the text file.

    Returns:
        dict: A dictionary where keys are terms and values are their frequencies.
    """
    terms = {}
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            match = re.match(r"(.+?): appears (\d+) time\(s\)\.", line.strip())
            if match:
                term, frequency = match.groups()
                terms[term] = int(frequency)
    return terms

# Function to compare terms between two files
def compare_term_frequencies(file_depth_1, file_depth_4):
    """
    Compares term frequencies between two files.

    Args:
        file_depth_1 (str): Path to the file for Depth 1.
        file_depth_4 (str): Path to the file for Depth 4.

    Returns:
        pd.DataFrame: A DataFrame containing terms and their frequencies in both files.
    """
    terms_depth_1 = read_terms_from_file(file_depth_1)
    terms_depth_4 = read_terms_from_file(file_depth_4)

    all_terms = set(terms_depth_1.keys()).union(set(terms_depth_4.keys()))  # Combine all unique terms
    comparison_data = []

    for term in all_terms:
        freq_depth_1 = terms_depth_1.get(term, 0)
        freq_depth_4 = terms_depth_4.get(term, 0)
        comparison_data.append({"Term": term, "Depth 1": freq_depth_1, "Depth 4": freq_depth_4})

    return pd.DataFrame(comparison_data)

# Function to visualize the comparison data
def plot_comparison(dataframe, title, save_path=None):
    """
    Visualizes the comparison of term frequencies using a bar chart.

    Args:
        dataframe (pd.DataFrame): DataFrame containing terms and their frequencies.
        title (str): Title for the chart.
    """
    # Display the comparison table
    print(f"\n{title}")
    print(dataframe.to_string(index=False))

    # Set the term column as the index for plotting
    dataframe = dataframe.set_index("Term")

    # Create a bar chart with logarithmic scale for frequencies
    ax = dataframe.plot(kind="bar", figsize=(12, 8), width=0.8, logy=True)
    plt.title(f"Frequency Comparison - {title}")
    plt.xlabel("Term")
    plt.ylabel("Frequency (Logarithmic Scale)")
    plt.xticks(rotation=45, ha="right")

    # Add frequency values above the bars
    for bar in ax.patches:
        height = bar.get_height()
        ax.annotate(f'{height:.0f}', (bar.get_x() + bar.get_width() / 2., height),
                    ha='center', va='center', fontsize=10, color='black', xytext=(0, 3), textcoords='offset points')

    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"Graph saved to {save_path}")

    plt.show()
    plt.close()

# File paths for Depth 1 and Depth 4
depth_1_files = [
    "basque/depth_1/terms_counter/terms_number_test_environmental_ba_en_v2.txt",
    "basque/depth_1/terms_counter/terms_number_test_environmental_ba_es_v2.txt",
    "basque/depth_1/terms_counter/terms_number_test_pharmasy_ba_en_v2.txt",
    "basque/depth_1/terms_counter/terms_number_test_pharmasy_ba_es_v2.txt",
]
depth_4_files = [
    "basque/depth_4/terms_counter/terms_number_environmental_ba_en_v2.txt",
    "basque/depth_4/terms_counter/terms_number_environmental_ba_es_v2.txt",
    "basque/depth_4/terms_counter/terms_number_pharmasy_ba_en_v2.txt",
    "basque/depth_4/terms_counter/terms_number_pharmasy_ba_es_v2.txt",
]

# Save paths for the comparison graphics
save_paths = [
    "basque/comparative_graphics_basque_wiki_depths/environmental_ba_en.png",
    "basque/comparative_graphics_basque_wiki_depths/environmental_ba_es.png",
    "basque/comparative_graphics_basque_wiki_depths/pharmacy_ba_en.png",
    "basque/comparative_graphics_basque_wiki_depths/pharmacy_ba_es.png",
]

# Titles for the comparisons
comparison_titles = [
    "Comparison of Environmental Terms (ba-en)",
    "Comparison of Environmental Terms (ba-es)",
    "Comparison of Pharmacy Terms (ba-en)",
    "Comparison of Pharmacy Terms (ba-es)"
]

# Compare each pair of files and visualize the results
for file_depth_1, file_depth_4, title, save_path in zip(depth_1_files, depth_4_files, comparison_titles, save_paths):
    comparison_df = compare_term_frequencies(file_depth_1, file_depth_4)
    plot_comparison(comparison_df, title, save_path=save_path)
```
